```python
#explicit type conversion
num_int = 123
num_str = "456"

print(type(num_int))
print(type(num_str))

num_str = int(num_str)
print(type(num_str))

num_sum = num_int + num_str

print("Sum of num_int and num_str:",num_sum)
print(type(num_sum))
```
